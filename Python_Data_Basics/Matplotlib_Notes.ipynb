```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Creating random data:
x_data = np.random.random(50) * 100
y_data = np.random.random(50) * 100
```

```python
# Scatterplot:
# plt.scatter(x,y, c="color", marker="marker", s=size(#), alpha=transparent(#))
plt.scatter(x_data, y_data, c="red", marker="*", s=100)
plt.show()
```

```python
# New Data
years = [2006 + x for x in range(16)]
print(years)
weights = [80, 83, 84, 85 ,86, 82, 81, 79, 83, 80,
           82, 82, 83, 81, 80, 79]
```

```python
# Line Chart:
# plt.plot(x, y, c="color", lw=line_width(#), linestyle="style") 
plt.plot(years, weights, c="red", lw=3, linestyle="--")
plt.show()
```

```python
# New Data:
x = ["C++", "C#", "Java", "Python", "Ruby"]
y = [30, 50, 20, 140, 10]
```

```python
# Bar Chart:
# plt.bar(x, y, color="color", edgecolor="color", width=width(#), lw=line_width(#))
plt.bar(x, y, color="red", edgecolor="black", width=0.5, lw=2)
plt.show()
```

```python
# New Data:
ages = np.random.normal(20, 3.5, 1000) # normal distribution with 20 mean, 1.5 STD, and 1000 values
```

```python
# Histogram Plot:
# plt.hist(data, bins=number, cumulative=boolean)
plt.hist(ages, bins=20, cumulative=True)
plt.show()
```

```python
# New Data:
langs = ["Python", "C++", "Java", "C#", "Ruby"]
votes = [50, 24, 14, 5, 17]
explodes = [0, 0, 0, 0.3, 0] # drags affect pieces out of the pie
```

```python
# Pie Chart:
plt.pie(votes, labels=langs, autopct="%1.1f%%", explode=explodes, startangle=90)
```

```python
# New Data:
heights = np.random.normal(172, 8, 300)
```

```python
# Box Plot:
plt.boxplot(heights)
plt.show()
```

### Plot Customization

```python
# New Data:
years = [2014, 2015, 2016, 2017,
         2018, 2019, 2020, 2021]
income = [55, 56, 62, 61,
          72, 72, 73, 75]
income_ticks = list(range(50,81,2))
```

```python
plt.plot(years, income)
plt.title("Income of John", fontsize=20) # sets title of plot
plt.xlabel("Year") # 
plt.ylabel("Yearly Income")
plt.yticks(income_ticks, [f"${x}k" for x in income_ticks])
plt.show()
```

```python
# New Data:
stock_a = [100, 102, 99, 101 , 101, 100, 102]
stock_b = [90, 95, 102, 104, 105, 103, 109]
stock_c = [110, 115, 100, 105, 100, 98, 95]
```

```python
# Legend:
plt.plot(stock_a, label="Company1")
plt.plot(stock_b, label="Company2")
plt.plot(stock_c, label="Company3")
plt.legend(loc="upper right")
plt.show()
```

```python
# Legend on Pie Chart + Style Sheets:
from matplotlib import style

style.use("dark_background")

votes = [10, 2, 5, 16, 22]
people = ['A', 'B', 'C', 'D', 'E']

plt.pie(votes, labels=None)
plt.legend(labels=people)
plt.show()
```

```python
# Sub plotting:
x = np.arange(100)

fig, axs = plt.subplots(nrows=2, ncols=2)

fig.suptitle("Four Plots")

axs[0, 0].plot(x, np.sin(x))
axs[0, 0].set_title("Sin")

axs[0, 1].plot(x, np.cos(x))
axs[0, 1].set_title("Cos")

axs[1, 0].plot(x, np.random.random(len(x)))
axs[1, 0].set_title("Random")

axs[1, 1].plot(x, np.log(x))
axs[1, 1].set_title("Logarithm")

plt.tight_layout()
plt.show()
```
